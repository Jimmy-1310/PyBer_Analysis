```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax=plt.subplots()
ax.plot(x_axis,y_axis,color="green",marker="D",linewidth=2,label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("Pyber Fare by Month")
ax.set_ylim(0,45)
ax.grid()
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis,x_axis,c="red",label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
y_axis_larger=[]

plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],c="skyblue",alpha=0.2,edgecolors="k",linewidth=2,label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%',colors=colors)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,8))
explode=(0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels=x_axis,autopct='%1.1f%%',shadow=True,startangle=90,counterclock=False,colors=colors,explode=explode)
plt.show()
```
